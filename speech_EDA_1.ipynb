### Filter and normalize
for sentence input, define function to: 
1. filter out uninteresting words
2. generalize remaining words
3. return word list

```python
from nltk.corpus import stopwords

def stop_n_lem(sentence):
    
    # import and define words to ignore for analysis
    stop = stopwords.words('english')
    stop += ['.', ',', '(', ')', "'", '"', ';', ':']
    stop = set(stop)
    
    # define lemmatizer
    lemma = nltk.stem.WordNetLemmatizer()
    
    words = word_tokenize(sentence) # split sentence into individual words
    words = [lemma.lemmatize(w) for w in words if w not in stop] # lemmatize non-stop words
    return(words)
```
